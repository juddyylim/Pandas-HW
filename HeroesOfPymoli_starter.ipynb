### Heroes Of Pymoli Data Analysis
* Of the 1163 active players, the vast majority are male (84%). There also exists, a smaller, but notable proportion of female players (14%).

* Our peak age demographic falls between 20-24 (44.8%) with secondary groups falling between 15-19 (18.60%) and 25-29 (13.4%).

```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# Raw data file
file_to_load = "Resources/purchase_data.csv"

# Read purchasing file and store into pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
total_players = len(purchase_data["SN"].unique())
total_players 
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
item = len(purchase_data["Item ID"].unique())
purchase = purchase_data["Purchase ID"].count()
average_price1 = round(sum(purchase_data["Price"]) / len(purchase_data["Price"]), 2)
average_price = '${:,.2f}'.format(average_price1)
total1 = purchase_data["Price"].sum()
total = '${:,.2f}'.format(total1)
summary = pd.DataFrame({"Number of Unique Items": [item],
                       "Average Price": [average_price],
                       "Number of Purchases": [purchase], 
                       "Total Revenue": [total]})
summary
```

## Gender Demographics

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
gender = purchase_data.groupby(['Gender'])
total_players1 = gender["SN"].nunique()
percentage = round((total_players1/total_players)*100, 2)
gd = pd.DataFrame({"Total Count": total_players1,
                  "Percentage of Players": percentage})
gd
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, etc. by gender


* For normalized purchasing, divide total purchase value by purchase count, by gender


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
gender = purchase_data.groupby(['Gender'])
gender.count().head()
pc = gender["Item Name"].count()
price = gender["Price"].mean()
price1 = round(gender["Price"].mean(), 2)
average = pc * price
average_total = round((average / total_players1), 2)
pa = pd.DataFrame({"Purchase Count": pc, 
                  "Average Purchase Price": price1,
                  "Total Purchase Value": average,
                  "Average Total Purchase per Person": average_total})
pa["Average Purchase Price"] = pa["Average Purchase Price"].map("${:.2f}".format)
pa["Total Purchase Value"] = pa["Total Purchase Value"].map("${:.2f}".format)
pa["Average Total Purchase per Person"] = pa["Average Total Purchase per Person"].map("${:.2f}".format)
pa
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Establish bins for ages
age_bins = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99999]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
purchase_data["Age_Group"] = pd.cut(purchase_data["Age"], age_bins, labels=group_names)
```

```python
age1 = purchase_data.groupby(["Age_Group"])
total_person = age1["SN"].nunique()
binpercentage = round((total_person/total_players)*100, 2)
demographics = pd.DataFrame({"Percentage of Players": binpercentage,
                            "Total Count": total_person})
demographics
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, etc. in the table below


* Calculate Normalized Purchasing


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
age1 = purchase_data.groupby(["Age_Group"])
age1.count().head()
binpc = age1["Item Name"].count()
binprice = age1["Price"].mean()
binprice1 = round(age1["Price"].mean(), 2)
total = binpc * binprice
total_average = round((total / total_person), 2)
total_average
age = pd.DataFrame({"Purchase Count": binpc,
                   "Average Purchase Price": binprice1,
                   "Total Purchase Value": total,
                   "Avg Total Purchase per Person": total_average})
age["Average Purchase Price"] = age["Average Purchase Price"].map("${:.2f}".format)
age["Total Purchase Value"] = age["Total Purchase Value"].map("${:.2f}".format)
age["Avg Total Purchase per Person"] = age["Avg Total Purchase per Person"].map("${:.2f}".format)
age
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
ts = purchase_data.groupby(["SN"])
tscount = ts["Item Name"].count()
tsaverage = ts["Price"].mean()
tsaverage1 = round(ts["Price"].mean(), 2)
tstotal = tscount * tsaverage
topspender = pd.DataFrame({"Purchase Count": tscount,
                          "Average Purchase Price": tsaverage1,
                          "Total Purchase Value": tstotal})
topsorted = topspender.sort_values(["Total Purchase Value"], ascending=False)
topsorted["Average Purchase Price"] = topsorted["Average Purchase Price"].map("${:.2f}".format)
topsorted["Total Purchase Value"] = topsorted["Total Purchase Value"].map("${:.2f}".format)
topsorted.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
new1 = purchase_data[["Item ID", "Item Name", "Price"]]
new1
new = purchase_data.groupby(["Item ID", "Item Name"])
newcount = new["Item Name"].count()
newprice = new["Price"].mean()
newtotal = newcount*newprice
popular = pd.DataFrame({"Purchase Count": newcount,
                       "Item Price": newprice,
                       "Total Purchase Value": newtotal})
popsorted = popular.sort_values(["Purchase Count"], ascending=False)
popsorted["Item Price"] = popsorted["Item Price"].map("${:.2f}".format)
popsorted["Total Purchase Value"] = popsorted["Total Purchase Value"].map("${:.2f}".format)
popsorted.head()
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
new1 = purchase_data[["Item ID", "Item Name", "Price"]]
new1
new = purchase_data.groupby(["Item ID", "Item Name"])
newcount = new["Item Name"].count()
newprice = new["Price"].mean()
newtotal = newcount*newprice
popular = pd.DataFrame({"Purchase Count": newcount,
                       "Item Price": newprice,
                       "Total Purchase Value": newtotal})
popsorted = popular.sort_values(["Total Purchase Value"], ascending=False)
popsorted["Item Price"] = popsorted["Item Price"].map("${:.2f}".format)
popsorted["Total Purchase Value"] = popsorted["Total Purchase Value"].map("${:.2f}".format)
popsorted.head()
```
